```python
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
import math
```

# One-D Gaussian

*Plotting Gaussian using Histogram of points sampled*
- *A 1D gaussian lies on a single dimension, but we need to plot it in 2D to visualize the density of points*
- *Histogram tells us the natural density*

```python
mu = np.array([0])
sigma = np.array([0.1]).reshape((1,1))
variance = (sigma**2)
# sigma = sigma.
```

```python
num_samples = 100000
one_d_gaussian_samples = np.random.multivariate_normal(mu, variance, num_samples)
```

```python
plot = plt.hist(one_d_gaussian_samples, bins=100)
```

*Plotting Probability Distribution*
- *Height is the value of f(x)*
- *This gives us the probability that the value is x. The integral of the curve over x is equal to 1*

\begin{equation*}
f(x)= {\frac{1}{\sigma\sqrt{2\pi}}}e^{- {\frac {1}{2}} (\frac {x-\mu}{\sigma})^2}
\end{equation*}

```python
def one_d_gaussian_prob_dist(x, mu, sigma):
    return (1/(sigma*((2*math.pi)**(1/2)))*(math.exp((-1/2)*((x-mu)/sigma)**2)))
```

```python
x = np.linspace(-1,+1,700)
y = [one_d_gaussian_prob_dist(x, 0, 0.1) for x in x]
plt.xlim([-0.4, +0.4])
plot = plt.scatter(x,y)
```

# Two-D Gaussian

```python
# Mean vector and covariance matrix
mu = np.array([0., 0.])
Sigma = np.array([[ 1. , .9], [.9,  1.]])
```

```python
two_d_gaussian_1 = np.random.multivariate_normal(mu, Sigma, 10000)
```

```python
x = two_d_gaussian_1[:,0]
y = two_d_gaussian_1[:,1]
```

```python
plt.scatter(x,y)
```
